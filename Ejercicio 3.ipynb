```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Datos
act_scores = np.array([28, 29, 25, 28, 28, 27, 27, 29, 21, 33, 28, 28, 28, 29, 26, 23, 30, 27, 24])
gpa = np.array([3.84, 3.75, 3.21, 3.65, 3.23, 3.87, 3.63, 3.75, 1.66, 3.20, 3.12, 3.41, 2.96, 3.38, 2.92, 3.53, 3.10, 2.03, 2.81])

# Ajuste de la recta por mínimos cuadrados
coef_gpa = np.polyfit(act_scores, gpa, 1)
poly_gpa = np.poly1d(coef_gpa)

# Gráfica
plt.scatter(act_scores, gpa, label='Datos')
plt.plot(act_scores, poly_gpa(act_scores), label='Recta de ajuste', color='red')
plt.legend()
plt.xlabel('Puntuación ACT')
plt.ylabel('Promedio de puntos')
plt.title('Ajuste de recta por mínimos cuadrados')
plt.show()

print(f"Coeficientes: {coef_gpa}")
```

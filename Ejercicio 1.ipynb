```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

xi = [4.0, 4.2, 4.5, 4.7, 5.1, 5.5, 5.9, 6.3, 6.8, 7.1]
yi = [102.56, 130.11, 113.18, 142.05, 167.53, 195.14, 224.87, 256.73, 299.50, 326.72]

# Ejercicio 1a.

# Ajuste de polinomio de grado 1
coef_1 = np.polyfit(xi, yi, 1)
poly_1 = np.poly1d(coef_1)

# Error cuadrático medio
error_1 = np.mean((poly_1(xi) - yi) ** 2)

# Gráfica
plt.scatter(xi, yi, label='Datos')
plt.plot(xi, poly_1(xi), label='Polinomio grado 1', color='red')
plt.legend()
plt.xlabel('xi')
plt.ylabel('yi')
plt.title('Ajuste de polinomio de grado 1')
plt.show()

print(f"Coeficientes del polinomio de grado 1: {coef_1}")
print(f"Error del polinomio de grado 1: {error_1}")


# Ejercicio 1b.
# Ajuste de polinomio de grado 2
coef_2 = np.polyfit(xi, yi, 2)
poly_2 = np.poly1d(coef_2)

# Error cuadrático medio
error_2 = np.mean((poly_2(xi) - yi) ** 2)

# Gráfica
plt.scatter(xi, yi, label='Datos')
plt.plot(xi, poly_2(xi), label='Polinomio grado 2', color='green')
plt.legend()
plt.xlabel('xi')
plt.ylabel('yi')
plt.title('Ajuste de polinomio de grado 2')
plt.show()

print(f"Coeficientes del polinomio de grado 2: {coef_2}")
print(f"Error del polinomio de grado 2: {error_2}")

# Ejercicio 1c.
# Ajuste de polinomio de grado 3
coef_3 = np.polyfit(xi, yi, 3)
poly_3 = np.poly1d(coef_3)

# Error cuadrático medio
error_3 = np.mean((poly_3(xi) - yi) ** 2)

# Gráfica
plt.scatter(xi, yi, label='Datos')
plt.plot(xi, poly_3(xi), label='Polinomio grado 3', color='blue')
plt.legend()
plt.xlabel('xi')
plt.ylabel('yi')
plt.title('Ajuste de polinomio de grado 3')
plt.show()

print(f"Coeficientes del polinomio de grado 3: {coef_3}")
print(f"Error del polinomio de grado 3: {error_3}")

#Ejercicio 1c.
# Datos
xi = np.array([4.0, 4.2, 4.5, 4.7, 5.1, 5.5, 5.9, 6.3, 6.8, 7.1])
yi = np.array([102.56, 130.11, 113.18, 142.05, 167.53, 195.14, 224.87, 256.73, 299.50, 326.72])

# Definición de la función
def func_exp(x, b, a):
    return b * np.exp(a * x)

# Ajuste
params_exp, _ = curve_fit(func_exp, xi, yi)

# Error cuadrático medio
error_exp = np.mean((func_exp(xi, *params_exp) - yi) ** 2)

# Gráfica
plt.scatter(xi, yi, label='Datos')
plt.plot(xi, func_exp(xi, *params_exp), label='Ajuste exponencial', color='purple')
plt.legend()
plt.xlabel('xi')
plt.ylabel('yi')
plt.title('Ajuste exponencial')
plt.show()

print(f"Parámetros de la forma y=be^ax: {params_exp}")
print(f"Error de la forma y=be^ax: {error_exp}")


#Ejercicio 1d.
# Definición de la función
def func_pow(x, b, a):
    return b * x ** a

# Ajuste
params_pow, _ = curve_fit(func_pow, xi, yi)

# Error cuadrático medio
error_pow = np.mean((func_pow(xi, *params_pow) - yi) ** 2)

# Gráfica
plt.scatter(xi, yi, label='Datos')
plt.plot(xi, func_pow(xi, *params_pow), label='Ajuste potencia', color='orange')
plt.legend()
plt.xlabel('xi')
plt.ylabel('yi')
plt.title('Ajuste potencia')
plt.show()

print(f"Parámetros de la forma y=bx^a: {params_pow}")
print(f"Error de la forma y=bx^a: {error_pow}")
```

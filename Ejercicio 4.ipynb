```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Datos
weights = np.array([4800, 3700, 3400, 2800, 1900])
percentages = np.array([3.1, 4.0, 5.2, 6.4, 9.6])

# Ajuste de la recta por mínimos cuadrados
coef_weights = np.polyfit(weights, percentages, 1)
poly_weights = np.poly1d(coef_weights)

# Gráfica
plt.scatter(weights, percentages, label='Datos')
plt.plot(weights, poly_weights(weights), label='Recta de ajuste', color='blue')
plt.legend()
plt.xlabel('Peso promedio (lb)')
plt.ylabel('Porcentaje de presentación')
plt.title('Ajuste de recta por mínimos cuadrados')
plt.show()

print(f"Coeficientes: {coef_weights}")
```
